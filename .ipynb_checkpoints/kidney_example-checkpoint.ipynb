## Chronic Kidney Disease Prediction

This tutorial illustrates the workflow of a data science project and how to use several important Python libraries for Exploratory Data Analysis and Modeling.The dataset is taken over 2-month period in India. It has 400 rows with 17 features, and the aim is to predict the presence or absence of chronic kidney disease using the patient examination. The actual result is based on a attribute named 'classification' which is either 'ckd'(chronic kidney disease) or 'notckd. The representation of all features are listed in the following table.

| Feature | Variable      | Value Type |
|---------|--------------|---------------|
| Age| age | numerical (years) |
| Blood pressure| bp | numerical (mm/Hg) |
| Specific Gravity| sg | nominal (1.005,1.010,1.015,1.020,1.025) |
| Albumin| al | nominal (0,1,2,3,4,5) |
| Sugar| su | nominal (0,1,2,3,4,5) |
| Pus Cell clumps| pcc | nominal (present,notpresent) |
| Bacteria| ba | nominal (present,notpresent) |
| Blood Glucose Random| bgr | numerical (mgs/dl) |
| Blood Urea| bu | numerical (mgs/dl) |
| Serum Creatinine| sc | numerical (mgs/dl) |
| Hemoglobin| hemo | numerical (gms) |
| Hypertension| htn | nominal (yes,no) |
| Diabetes Mellitus| dm | nominal (yes,no) |
| Coronary Artery Disease| cad | nominal (yes,no) |
| Appetite| appet | nominal (good,poor) |
| Pedal Edema| pe | nominal (yes,no) |
| Anemia| ane | nominal (yes,no) |
| Presence or absence of kidney disease|classification | nominal (ckd,notckd) |


The dataset was split into kidney_disease_train.csv and kidney_disease_test.csv.

The workflow is as follows:
* 1. Importing data from csv files to Pandas dataframes
* 2. Inspecting data with Pandas
* 3. Exploratory Data Analysis
* 4. Featuring and cleaning data if needed
* 5. Modeling
* 6. Evaluation

### 5. Cleaning and Featuring
- Cleaning involves dealing with missing values. If the number of missing values is small, we fill them with the mode for categorical features and with the mean for numerical features

- Featuring involves transforming data: Converting categorical values to numerical values

```python
# Counting the number of missing values in each column of train set
df_train.isnull().sum()
```

```python
# Counting the number of missing values in each column of test set
df_test.isnull().sum()
```

```python
# Drop id column in both sets 
df_train.drop('id',axis=1,inplace=True)
df_test.drop('id',axis=1,inplace=True)
```

```python
# Filling missing values 
num_features = ['age', 'bp', 'bgr', 'bu', 'sc', 'hemo']
for col in num_features:
    df_train[col].fillna(df_train[col].mean(), inplace=True)
    df_test[col].fillna(df_train[col].mean(), inplace=True)
    
cat_features = ['sg', 'al', 'su', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'classification']
for col in cat_features:
    df_train[col].fillna(df_train[col].mode().iloc[0], inplace=True)
    df_test[col].fillna(df_train[col].mode().iloc[0], inplace=True)
```

```python
# Check missing value again
print('Number of missing values in train set:', df_train.isnull().sum().sum())
print('Number of missing values in test set:', df_test.isnull().sum().sum())
```

### Featuring
Convert categorical values to numerical values

```python
df_train['pcc'] = df_train['pcc'].map({'notpresent':0, 'present': 1})
df_test['pcc'] = df_test['pcc'].map({'notpresent':0, 'present': 1})
```

```python
df_train['ba'] = df_train['ba'].map({'notpresent':0, 'present': 1})
df_test['ba'] = df_test['ba'].map({'notpresent':0, 'present': 1})
```

```python
df_train['htn'] = df_train['htn'].map({'no':0, 'yes': 1})
df_test['htn'] = df_test['htn'].map({'no':0, 'yes': 1})
```

```python
df_train['dm'] = df_train['dm'].map({'no':0, 'yes': 1})
df_test['dm'] = df_test['dm'].map({'no':0, 'yes': 1})
```

```python
df_train['cad'] = df_train['cad'].map({'no':0, 'yes': 1})
df_test['cad'] = df_test['cad'].map({'no':0, 'yes': 1})
```

```python
df_train['appet'] = df_train['appet'].map({'poor':0, 'good': 1})
df_test['appet'] = df_test['appet'].map({'poor':0, 'good': 1})
```

```python
df_train['pe'] = df_train['pe'].map({'no':0, 'yes': 1})
df_test['pe'] = df_test['pe'].map({'no':0, 'yes': 1})
```

```python
df_train['ane'] = df_train['ane'].map({'no':0, 'yes': 1})
df_test['ane'] = df_test['ane'].map({'no':0, 'yes': 1})
```

```python
df_train['classification'] = df_train['classification'].map({'notckd':0, 'ckd':1})
df_test['classification'] = df_test['classification'].map({'notckd':0, 'ckd':1})
```

```python
# Separate feature columns and label columns for training and evaluating
X_train = df_train.drop('classification', axis=1) # features
y_train = df_train['classification']            # labels

X_test = df_test.drop('classification', axis=1)
y_test = df_test['classification']
```

## 5-6. Modeling and Evaluating



```python
# Import APIs from scikit-learn
from sklearn.preprocessing import MinMaxScaler, StandardScaler

from sklearn.neighbors import KNeighborsClassifier  #import k-Nearest Neighbor model from scikit-learn library 
from sklearn.linear_model import LogisticRegression #import Logistic Regression model from scikit-learn library 
from sklearn.tree import DecisionTreeClassifier     #import Decision Tree model from scikit-learn library 

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix

import seaborn as sns
```

Scaling features 

```python
# Scale features 
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### K-Nearest Neighbor Model
Varying the number of neighbors (n_neighbors) to obtain the best performance

```python
# Call the K-nearest model
KNN = KNeighborsClassifier(n_neighbors=5)   # Adjust n_neighbors to see the performance

# Fit the model on train set
KNN.fit(X_train_scaled, y_train)           

# Use the trained model to predict the outcomes of test set
y_predict_knn = KNN.predict(X_test_scaled)

# Compute the evaluation metrics
print('Accuracy: {}'.format(accuracy_score(y_test, y_predict_knn)))
print('Precision: {}'.format(precision_score(y_test, y_predict_knn)))
print('Recall: {}'.format(recall_score(y_test, y_predict_knn)))
```

Visualizing confusion matrix

```python
cm_knn = confusion_matrix(y_test, y_predict_knn)
                          
# Transform to df for easier plotting
cm_knn_df = pd.DataFrame(cm_knn,
                     index = ['notckd','ckd'], 
                     columns = ['notckd','ckd'])

plt.figure(figsize=(6,4))
sns.heatmap(cm_knn_df, annot=True)
plt.xlabel("Predict")
plt.ylabel("True")
```

### Logistic Regression Model

```python
# Call the Logistic Regression model
LR = LogisticRegression()  

# Fit the model on train set
LR.fit(X_train_scaled, y_train)            # 

# Use the trained model to predict the outcomes of test set
y_predict_lr = LR.predict(X_test_scaled)

# Compute the evaluation metrics
print('Accuracy: {}'.format(accuracy_score(y_test, y_predict_lr)))
print('Precision: {}'.format(precision_score(y_test, y_predict_lr)))
print('Recall: {}'.format(recall_score(y_test, y_predict_lr)))
```

Visualizing confusion matrix

```python
cm_lr = confusion_matrix(y_test, y_predict_lr)
                          
# Transform to df for easier plotting
cm_lr_df = pd.DataFrame(cm_lr,
                     index = ['notckd','ckd'], 
                     columns = ['notckd','ckd'])

plt.figure(figsize=(6,4))
sns.heatmap(cm_lr_df, annot=True)
plt.xlabel("Predict")
plt.ylabel("True")
```

###  Decision Tree Model

```python
# Call the Decision Tree model
DT = DecisionTreeClassifier() 

# Fit the model on train set. It is suggested to train the model on the unscaled train set.
DT.fit(X_train, y_train)  

# Use the trained model to predict the outcomes of test set
y_predict_dt = DT.predict(X_test)

# Compute the evaluation metrics
print('Accuracy: {}'.format(accuracy_score(y_test, y_predict_dt)))
print('Precision: {}'.format(precision_score(y_test, y_predict_dt)))
print('Recall: {}'.format(recall_score(y_test, y_predict_dt)))
```

Visualizing confusion matrix

```python
cm_dt = confusion_matrix(y_test, y_predict_dt)
                          
# Transform to df for easier plotting
cm_dt_df = pd.DataFrame(cm_dt,
                     index = ['notckd','ckd'], 
                     columns = ['notckd','ckd'])

plt.figure(figsize=(6,4))
sns.heatmap(cm_dt_df, annot=True)
plt.xlabel("Predict")
plt.ylabel("True")
```
